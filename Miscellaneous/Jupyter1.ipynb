```python
from numpy import linspace
import matplotlib.pyplot as plt
from math import floor, sqrt

def greatestIntegerFunction(number):
    return int(floor(number))
plt.plot([j for j in linspace(-.1,0.001,1000)],[greatestIntegerFunction(1/i) for i in linspace(-.1,0.001,1000)])
plt.show()
```

x²/a² + y²/b² = 1

y = ± sqrt(b² * (1 - x²/a²))

```python
a = 25
b = 20
max_coordinate = max(a,b) + 1
sensitivity = 1000
def ellipsePlot(x):
    return sqrt((b**2) * (1 - (x**2)/(a**2)))
fig, ax = plt.subplots()
ax.plot([i for i in linspace(-a,a,sensitivity)],[ellipsePlot(i) for i in linspace(-a,a,sensitivity)])
ax.plot([i for i in linspace(-a,a,sensitivity)],[-ellipsePlot(i) for i in linspace(-a,a,sensitivity)])
```
